```julia
using Plots

# X is a matrix, y is an array of points
X, y = make_blobs(n_samples=200)
print(y)
plot(X[:, 1], X[:, 2], seriestype=:scatter, marker_z = y, legend=false)
```

```julia
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

```julia
plot(X[:, 1], X[:, 2], seriestype=:scatter, marker_z=y_kmeans, legend=false)
```

```julia
using LinearAlgebra

A = kneighbors_graph(X, n_neighbors=10, mode="connectivity").toarray()
```

```julia
using LinearAlgebra

D = np.diag(np.sum(A, axis=1))
L = D - A
```

```julia
using LinearAlgebra
using DataFrames

k = 2

eigvalues = np.real(eigvals(L))
sorted_e_values = sort(eigvalues)

print(sorted_e_values)

eigvectors = np.real(eigvecs(L))[:,1:3]

df1 = eigvectors
```

```julia
k_means = KMeans(n_clusters=3)

k_means.fit(df1)
```

```julia
using Plots

plot(X[:, 1], X[:, 2], seriestype=:scatter, marker_z = k_means.labels_, color =:lightrainbow, legend=false)
```
